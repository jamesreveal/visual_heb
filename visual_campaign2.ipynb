```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])


df = pd.read_csv("dashboard_export - HEB Report on Thu Dec 08 2022 11_04_09 GMT-0800 (Pacific Standard Time).csv")
n_columns=len(df.columns)

results=[]
labels=[]
for k in range(n_columns):
    result=0
    n = len(df)
    for i in range(n):
        item = df.iloc[:,k][i]
        if isinstance(item,str) and len(item)>0:
            result+=1 

    results.append(result)
    labels.append(df.columns[k])




# creating the dataset

fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(labels, results, color ='maroon',
        width = 0.4)

addlabels(labels, results)
 
plt.xlabel("Questions asked")
plt.ylabel("No. of persons responding")
plt.title("HEB Campaign Statisitcs")
plt.xticks(rotation = 45)
plt.show()



```
